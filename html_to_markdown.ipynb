## Import packages

```python
import advertools as adv
import pandas as pd
```

## Crawl a website

```python
adv.crawl(url_list="https://searchnstuff.co.uk/", output_file="sns_crawl.jsonl", follow_links=True)
```

## Read the file into a DataFrame (table)

```python
crawldf = pd.read_json("sns_crawl.jsonl", lines=True)
crawldf.head()
```

## Convert pages' content to markdown

```python
crawldf['markdown'] = adv.crawlytics.generate_markdown(crawldf)
```

## Done!

## How does it work?

- Take the text in the `body_text` column
- Take all the avilable headings h1-h6
- For each heading, use a regex to locate that heading in `body_text`
- Partition `body_text` on the locations of all the found headings
- Return a list of markdown strings

It's usually good to save these as an additional columns in the crawl DataFrame for easier processing

```python
crawldf.filter(regex=r'h\d|body_text').head(2)
```
